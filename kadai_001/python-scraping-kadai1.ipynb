```python
import requests
from bs4 import BeautifulSoup
```

```python
# タイトルを取得する対象のUR
url = 'https://news.yahoo.co.jp/articles/1b733d2a4ecf0e6a3dcaca4b066a0aa9c908b09c'
```

```python
# GETリクエストを送信し、HTMLを取得する
response = requests.get(url)
```

```python
# HTMLを解析する
soup = BeautifulSoup(response.text,'html.parser')
```

```python
# 本文の要素をCSSセレクタを使用して取得する
article_element = soup.select_one('#uamods > div.article_body.highLightSearchTarget > div > p')
# 本文のテキストを取得する
article_text = article_element.text
```
